# Preliminary analysis of RCA in vivo trials

```python
import os, sys
import tifffile as tif
import numpy as np
import matplotlib.pyplot as plt
import tqdm
```

## Crop and save images

```python
data_dir = '/mnt/transfer_nas/Sildescanner_write/Quentin//20250404_eGFP_RCA_Fluo'
cropped_dir = os.path.join(data_dir, 'cropped')

# list of all files in the directory
files = os.listdir(data_dir)
files = [os.path.join(data_dir, f) for f in files if f.endswith('.tif')]
files.sort()
files
```

```python
# read in files and crop the 1000x1000 pixels in the center
def crop_center(image, crop_size):
    h, w = image.shape
    start_h = (h - crop_size) // 2
    start_w = (w - crop_size) // 2
    return image[start_h:start_h + crop_size, start_w:start_w + crop_size]

def crop_and_save_tiffs(files, crop_size=1000):
    for file in tqdm.tqdm(files):
        # read the tiff file
        image = tif.imread(file)
        
        # crop the image
        cropped_image = crop_center(image, crop_size)
        
        # save the cropped image
        cropped_file = os.path.join(cropped_dir, os.path.basename(file))
        tif.imwrite(cropped_file, cropped_image)
        print(f'Saved cropped image to {cropped_file}')
```

```python
crop_and_save_tiffs(files, crop_size=2000)
```

## Visualized cropped image

```python
def plot_rgb_image(image, rlim, glim, blim):
    # Create a figure and axis
    fig, ax = plt.subplots(figsize=(10, 10))
    
    # Plot the RGB image
    ax.imshow(image)
    
    # Hide the axes
    ax.axis('off')
    
    # Show the plot
    plt.show()
    
def plot_gray_image(image, title, rlim, glim, blim):
    # Create a figure and axis
    fig, ax = plt.subplots(1, 3, figsize=(12, 5), dpi = 200)
    
    # Plot the red channel
    ax[0].imshow(image[:, :, 0], 'gray', vmin=rlim[0], vmax=rlim[1])
    ax[0].set_title('550 nm (Red)')
    ax[0].axis('off')
    # Plot the green channel
    ax[1].imshow(image[:, :, 1], 'gray', vmin=glim[0], vmax=glim[1])
    ax[1].set_title('488 nm (Green)')
    ax[1].axis('off')
    # Plot the blue channel
    ax[2].imshow(image[:, :, 2], 'gray', vmin=blim[0], vmax=blim[1])
    ax[2].set_title('405 nm (Blue)')
    ax[2].axis('off')
    # Set the title
    fig.suptitle(title, fontsize=16)
    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
# Read the cropped images and plot them
cropped_files = os.listdir(cropped_dir)
cropped_files = [os.path.join(cropped_dir, f) for f in cropped_files if f.endswith('.tif')]
cropped_files.sort()

cond_dict = {
    'WT_standard': 'S1',
    'WT_50mix': 'S2',
    'eGFP_standard': 'S3',
    'eGFP_50mix': 'S4'
}

color_list = ['Cy3', 'EGFP', 'DAPI']

# read in the image from the selected condition
def get_img_name_condition_color(cropped_files, cond, color):
    for file in cropped_files:
        if cond in file and color in file:
            return file
    return None

def read_img(cropped_files, cond, color):
    img_r = []
    img_g = []
    img_b = []
    for color, idx in zip(color_list, range(3)):
        img_name = get_img_name_condition_color(cropped_files, cond, color)
        if img_name is not None:
            img = tif.imread(img_name)
            # normalize the image to 0-255
            img = (img - 0) / (img.max() - 0) * 255
            img = img.astype(np.uint8)
            if idx == 0:
                img_r = img
            elif idx == 1:
                img_g = img
            elif idx == 2:
                img_b = img
    # stack the images into a 3D array [channel, height, width]
    img_rgb = np.stack((img_r, img_g, img_b), axis=-1)
    return img_rgb
```

```python
cond = 'eGFP_50mix'
img_rgb = read_img(cropped_files, cond_dict[cond], color_list)
img_rgb.shape
```

```python
rlim = (40, 80)
glim = (0, 100)
blim = (0, 200)
plot_gray_image(img_rgb, cond, rlim, glim, blim)
```
