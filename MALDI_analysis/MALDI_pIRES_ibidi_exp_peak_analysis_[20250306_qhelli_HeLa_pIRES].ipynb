## Image visualization

```python
def get_intensity(spectra, mz, tolerance):
    """
    spectra: pandas series, the spectra data
    mz: float, the mz value
    tolerance: float, the tolerance value, plus/minus the target mz
    """
    # Get the mz values within the tolerance range
    mz_range = (mz - tolerance, mz + tolerance)
    
    # Filter the spectra to get the intensity at the target mz
    intensity = spectra[(spectra.index >= mz_range[0]) & (spectra.index <= mz_range[1])]
    
    # Return the sum of intensities within the range
    return intensity.sum()

def plot_mz_map(cond, mz, tolerance, pdata_df, sdata_df, colorlim=None, xyscale = None):
    """
    Plot the mz map for the given condition.
    cond: str, the condition to plot, corresponding to the "run" column in pdata_df
    mz: float, the mz value
    tolerance: float, the tolerance value, plus/minus the target mz
    pdata_df: pandas DataFrame, the dataframe with pixel metadata
    sdata_df: pandas DataFrame, the dataframe with spectra data
    colorlim: list of float, the color limits for the plot (min, max)
    xyscale: list of int, the x and y range for the image (for better alignment)
    """
    
    pdata_df_cond = pdata_df[pdata_df["run"] == cond]
    sdata_df_cond = sdata_df.loc[pdata_df_cond.index]
    
    # Get the intensity for each pixel and generate a array
    x_range = pdata_df_cond["x"].max()+1
    y_range = pdata_df_cond["y"].max()+1
    intensity_map = np.zeros((y_range, x_range))
    
    for i in range(len(pdata_df_cond)):
        x = pdata_df_cond.iloc[i]["x"]
        y = pdata_df_cond.iloc[i]["y"]
        spectra = sdata_df_cond.iloc[i]
        
        intensity = get_intensity(spectra, mz, tolerance)
        
        intensity_map[y, x] = intensity
    # Plot the intensity map
    plt.imshow(intensity_map, cmap='gray', interpolation='nearest')
    # plt.colorbar(label='Intensity')
    # plt.title(f'MZ: {mz} Da, Condition: {cond}')
    # plt.xlabel('X Coordinate')
    # plt.ylabel('Y Coordinate')
    plt.axis('off')
    
    if colorlim is not None:
        plt.clim(colorlim[0], colorlim[1])
        
    if xyscale is not None:
        plt.xlim(0, xyscale[0])
        plt.ylim(0, xyscale[1])
    
def plot_mz_map_multi(cond, mz, tolerance, pdata_df, sdata_df, colorlim=None, xyscale = None):
    """
    Plot the mz map for the given condition.
    cond: list of str, the conditions to plot, corresponding to the "run" column in pdata_df
    mz: list of float, the mz values
    tolerance: float, the tolerance value, plus/minus the target mz
    pdata_df: pandas DataFrame, the dataframe with pixel metadata
    sdata_df: pandas DataFrame, the dataframe with spectra data
    """
    
    # check how many conditions there are in total
    num_conditions = len(cond) * len(mz)
    # set up the figure
    plt.figure(figsize=(7, 3.62), dpi = 200)
    
    # loop through the conditions and mz values
    for j, mz_value in enumerate(mz):
        for i, condition in enumerate(cond):
            plt.subplot(len(mz), len(cond), j * len(cond) + i + 1)
            plot_mz_map(condition, mz_value, tolerance, pdata_df, sdata_df, colorlim, xyscale)
            
            if j == 0:
                cond_title = condition.split("_")[:-1]
                # remove "pIRES" 
                if "pIRES" in cond_title:
                    cond_title.remove("pIRES")
                cond_title = " ".join(cond_title)
                plt.title(cond_title, fontsize=8)
                
            if i == 0:
                # add text to the left of the plot
                plt.text(-1, 15, f"{mz_value}±1 Da", fontsize=8, ha='right', va='center', rotation=0) 
                
    # set the colorbar, outside the plot
    # cbar = plt.colorbar(orientation='horizontal', pad=0.05)
    # cbar.set_label('Intensity', fontsize=8)

    
    
    plt.tight_layout(pad=0.2)
    plt.show()
    
```

```python
# Example usage
# Define the conditions and mz values to plot
conditions = ["pIRES_GFP_s", "pIRES_Amb_GFP_s", "pIRES_NWS_GFP_s", "pIRES_HA_GFP_s"]
mz_values = [1254.53, 1835.93, 1879.94]

tolerance = 1
# Plot the mz maps for the given conditions and mz values
plot_mz_map_multi(conditions, mz_values, tolerance, pdata_df, sdata_df, colorlim=[0, 15], xyscale=[40, 30])
    
```

## Spectra visualization

```python
# get the mean spectrum for each condition
sdata_df_mean = pd.DataFrame(index=conditions, columns=sdata_df.columns)
for cond in conditions:
    pdata_df_cond = pdata_df[pdata_df["run"] == cond]
    sdata_df_cond = sdata_df.loc[pdata_df_cond.index]
    
    # Get the mean spectrum for the condition
    sdata_df_mean.loc[cond] = sdata_df_cond.mean()
    
sdata_df_mean
```

```python
cdict = {
    "pIRES_GFP_s": "#D0D3D6",
    "pIRES_Amb_GFP_s": "#2366EA",
    "pIRES_NWS_GFP_s": "#C3D3F3",
    "pIRES_HA_GFP_s": "#38599B"
}

# plot the mean spectrum for each condition
# show only two range of mz: 1200-1300 and 1800-1900, make a break in the x axis
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(4, 2.5), dpi = 300)

# plot vertical lines
lw = 2
alp = 0.5
ax1.axvline(x=1254.53, color=cdict["pIRES_Amb_GFP_s"], linestyle='-', lw=lw+2, alpha = alp)
ax2.axvline(x=1835.93, color=cdict["pIRES_NWS_GFP_s"], linestyle='-', lw=lw+2, alpha = alp)
ax2.axvline(x=1879.94, color=cdict["pIRES_HA_GFP_s"], linestyle='-', lw=lw+2, alpha = alp)

# plot the first range
for cond in conditions:
    ax1.plot(sdata_df_mean.columns, sdata_df_mean.loc[cond], label=cond, color=cdict[cond], lw=lw, alpha = alp+0.2)
    ax2.plot(sdata_df_mean.columns, sdata_df_mean.loc[cond], label=cond, color=cdict[cond], lw=lw, alpha = alp+0.2)
    
ax1.set_xlim(1200, 1300)
ax1.set_ylim(0, 20)

ax2.set_xlim(1800, 1900)
ax2.set_ylim(0, 20)

ax1.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.yaxis.set_visible(False)

# add the diagonal line
d = 0.02
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False, lw=0.5)
ax1.plot((1-d, 1+d), (-d, +d), **kwargs)
ax1.plot((1-d, 1+d), (1-d, 1+d), **kwargs)
kwargs.update(transform=ax2.transAxes)
ax2.plot((-d, +d), (1-d, 1+d), **kwargs)
ax2.plot((-d, +d), (-d, +d), **kwargs)

ax1.set_xticks([1200, 1250, 1300])
ax2.set_xticks([1800, 1850, 1900])
ax1.set_yticks([0, 5, 10, 15, 20])

# set distance between subplots
plt.subplots_adjust(wspace=0.3)

# add x labels
# plot at the center bottom of the two subplots
ax2.text(1797, -4, "m/z", ha='right', va='center', fontsize=14)
ax1.set_ylabel("Intensity", fontsize=14)

# show legend on the top of the plot, outside the plot, horizontal
# define the legend text content
legend_text = ["GFP", "Amb GFP", "NWS GFP", "HA GFP"]
plt.legend(legend_text, loc='upper center', bbox_to_anchor=(-0.15, 1.5), ncol=4, fontsize=14, frameon=False, columnspacing=0.8)
plt.show()
```
