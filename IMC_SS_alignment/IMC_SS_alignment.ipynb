# Try to align down-sampled MALDI crater image and IMC panorama

```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt
import tifffile as tif
import SimpleITK as sitk
from skimage import restoration, util
```

```python
code_dir = os.path.dirname(os.getcwd())
test_data_dir = os.path.join(code_dir, "test-data")
```

```python
# read the image
pano = tif.imread(os.path.join(test_data_dir, "panorama_test.tiff"))
crater = tif.imread(os.path.join(test_data_dir, "crater_downsampled.tiff"))
```

```python
print(pano.shape)
print(crater.shape)
```

```python
# convert both image to grayscale
pano_gray = pano.mean(axis=2)[1000:3500, 3000:5000]
crater_gray = crater.mean(axis=2)
print(pano_gray.shape)
print(crater_gray.shape)
```

```python
# show the images
plt.figure()
plt.imshow(pano_gray, cmap='gray')
plt.colorbar()
plt.figure()
plt.imshow(crater_gray, cmap='gray')
plt.show()
```
